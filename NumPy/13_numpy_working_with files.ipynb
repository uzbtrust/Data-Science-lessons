```python
import numpy as np
```

```python
arr = np.arange(0,10)
```

```python
np.save("data.npy", arr)
```

```python
np.load('data.npy')
```

```python
arr1 = np.random.rand(2,6)
```

```python
np.savez("data1", a=arr, b=arr1)
```

```python
data1 = np.load("data1.npz")
data1
```

```python
data1['a']
```

```python
data1['b']
```

```python
np.savez_compressed("data2", a=arr, b=arr1)
data2 = np.load("data2.npz")
data2['a']
```
