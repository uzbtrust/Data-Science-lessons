```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
%matplotlib inline
```

Machine Learning

image.png

```python
#data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

plt.scatter(x_data, y_data)
plt.show()
```

Classic

```python
def classic_program(X): return X*70

y_pred= [classic_program(x) for x in x_data]

plt.scatter(x_data, y_data)
plt.plot(x_data, y_pred)
plt.show()
```

Weight assignment

```python
def weight_asigned_program(X, w): return w*X*70

w = 1.3
y_pred= [weight_asigned_program(x,w) for x in x_data]

```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression().fit(x_data, y_data)
y_predict = lr_model.predict(x_data)

equation = f'y = {lr_model.coef_[0]:.3f} * x + {lr_model.intercept_:.3f}'
print(equation)

plt.scatter(x_data, y_data)
plt.plot(x_data, y_predict)
plt.show()
```

Deep Learning

image.png

```python
from sklearn.neural_network import MLPRegressor

data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

mlp_nodel = MLPRegressor(max_iter=2000).fit(x_data, y_data)
y_predict = mlp_nodel.predict(x_data)

plt.scatter(x_data, y_data)
plt.plot(x_data, y_predict)
plt.show()
```
