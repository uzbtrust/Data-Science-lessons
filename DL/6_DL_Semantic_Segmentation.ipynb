```python
from fastai.vision.all import *
import numpy as np

path = untar_data(URLs.CAMVID_TINY)

# label getter
def label_gatter(file):
  return path/'labels'/f'{file.stem}_P{file.suffix}'

# dataloaders
dls = SegmentationDataLoaders.from_label_func(path, bs=8, fnames=get_image_files(path/'images'),
                                               label_func=label_gatter, codes=np.loadtxt(path/'codes.txt', dtype=str))

# fine-tuning
learn = unet_learner(dls, resnet34)
learn.fine_tune(12)
```

```python
learn.show_results(max_n=6, figsize=(15,16))
```
