```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("new.db")
cursor = db.cursor()
```

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop (
  rank INT,
  country TEXT PRIMARY KEY,
  density REAL,
  population DOUBLE NOT NULL,
  area INT NOT NULL
);
"""
cursor.execute(query)
db.commit()
```

```python
query = "INSERT INTO worldpop VALUES(1, 'Macaoe', 21946.4667, 658.3940, 30)"
cursor.execute(query)
db.commit()
```
