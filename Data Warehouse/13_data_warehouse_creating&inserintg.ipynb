```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("new.db")
cursor = db.cursor()
```

```python
df = pd.read_csv("drive/MyDrive/datasets/Electricity_Production_By_Source.csv")
df.columns = ['country', 'code', 'year', 'coal', 'gas', 'hydro', 'renewable_other', 'solar', 'oil', 'wind', 'nuclear']
df.head(3)
```

```python
df.to_sql("electricity", db)
```

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop (
  rank INT,
  country TEXT PRIMARY KEY,
  density REAL,
  population DOUBLE NOT NULL,
  area INT NOT NULL
);
"""
cursor.execute(query)
db.commit()
```
