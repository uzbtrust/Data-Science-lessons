```python
db = sql.connect('electricity.db')
cursor = db.cursor()
```

```python
dfuz = pd.read_sql('SELECT year, coal, gas, hydro FROM elect WHERE country="Uzbekistan"', db)
dfuz
```

```python
# BETWEEN
dfuz = pd.read_sql('SELECT year, coal, gas, hydro FROM elect WHERE country="Uzbekistan" AND year BETWEEN 2000 AND 2025', db)
dfuz
```

```python
# IN
dfuz = pd.read_sql('SELECT year, coal, gas, hydro FROM elect WHERE country="Uzbekistan" AND year IN (2000,2005,2010,2025)', db)
dfuz
```

```python
# LIKE
# %var -> Sarvar, Anvar
# _ar -> car, bar
# __ar -> bear, star
query = 'SELECT country FROM elect WHERE country LIKE "%stan"'
df = pd.read_sql(query, db)
df
```
