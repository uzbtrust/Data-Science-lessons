```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("new.db")
cursor = db.cursor()
```

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop (
  rank INT,
  country TEXT PRIMARY KEY,
  density REAL,
  population DOUBLE NOT NULL,
  area INT NOT NULL
);
"""
cursor.execute(query)
db.commit()
```

```python
df = pd.read_csv("drive/MyDrive/datasets/world_population.csv", index_col=0)
uz = df[df['country'] == 'Uzbekistan']
uz
```

```python
rank = uz.index.values[0]
country = uz.at[rank, 'country']
density = uz.at[rank, 'density']
population = uz.at[rank, 'pop2021']
area = uz.at[rank, 'area']
```

```python
cursor.execute(f"""
INSERT INTO worldpop VALUES ({rank}, '{country}', {density}, {population}, {area});
""")
db.commit()
```

```python
cursor.execute('SELECT * FROM worldpop').fetchall()
```
