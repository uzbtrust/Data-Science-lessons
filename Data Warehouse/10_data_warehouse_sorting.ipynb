```python
db = sql.connect("chinook.db")
cursor = db.cursor()
```

```python
cursor.execute('SELECT * FROM customers').fetchall()
command = 'SELECT * FROM customers'
df = pd.read_sql(command, db, index_col='CustomerId')
df.head()
```

```python
command = 'SELECT * FROM customers ORDER BY LastName'
df = pd.read_sql(command, db, index_col='CustomerId')
df.head()
```

```python
query = 'SELECT DISTINCT Country FROM customers'
df = pd.read_sql(query, db)
df
```
