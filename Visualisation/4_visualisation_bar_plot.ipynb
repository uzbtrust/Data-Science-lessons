```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv', index_col=0)
df.columns = ['maydon', 'aholi']
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
plt.title("O'zbekiston hududlardagi aholi")
plt.xlabel("Hududlar")
plt.ylabel("Ahollar soni")
plt.grid()
sns.barplot(x=df.index, y=df.aholi, data=df, palette="viridis")
plt.show()
```
