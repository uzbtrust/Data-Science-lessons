```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv', index_col=0)
df.head()
```

```python
df = df[df.year>2010]
plt.figure(figsize=(16,10))
sns.scatterplot(data=df, x=df['mileage'], y=df['price'], hue='transmission', style='fuelType', s=100)
plt.show()
```

```python
sns.swarmplot(data=df, x='transmission', y='price')
plt.show()
```
