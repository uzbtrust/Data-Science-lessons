```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv', index_col=0)
df.head()
```

```python
df = df[df.year>2010]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x=df['year'], y=df['price'])
sns.regplot(data=df, x=df.year, y=df.price)
plt.show()
```
