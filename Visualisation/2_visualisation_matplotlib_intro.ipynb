```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y, linewidth=2, linestyle="-", color="#FFD700", marker="o")
plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12,6))
plt.plot(x,sin,
         linewidth=1.5,
         linestyle="-.",
         color="#8E44AD",
         alpha=0.7,
         marker="h")
plt.plot(x, cos,
         linestyle=":",
         color="c",
         alpha=0.7,
         marker="v")
plt.title("six(x) va cos(x) funksiya grafigi")
plt.xlabel("pi")
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize="x-large")
plt.grid(color="tomato", alpha=0.5, linestyle="--")
plt.savefig("sin&cos.png")
plt.show()
```
