```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df = df[df.year>=2010]
df = df[df.transmission != 'Other']
df = df[df.fuelType != 'Other']
df.head()
```

```python
x = np.arange(0,4*np.pi,0.1)
sin = np.sin(x)
cos = np.cos(x)

fig1 = plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(x,sin,
         linewidth=1.5,
         linestyle="-.",
         color="#8E44AD",
         alpha=0.7,
         marker="h")
plt.subplot(2,1,2)
plt.plot(x, cos,
         linestyle=":",
         color="c",
         alpha=0.7,
         marker="v")
plt.title("six(x) va cos(x) funksiya grafigi")
plt.xlabel("pi")
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize="x-large")
plt.grid(color="tomato", alpha=0.5, linestyle="--")
plt.savefig("sin&cos.png")
plt.show()
```
