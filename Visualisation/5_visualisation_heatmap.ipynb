```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/museum_visitors.csv', index_col=0, parse_dates=[0], usecols=[0,1])
df.head()
```

```python
type(df.index)
```

```python
df['month'] = [i.month for i in df.index]
df.head()
```

```python
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['month', 'year']).max()
df2.head()
```

```python
df2.unstack(level=1)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data=df2.unstack(level=0), cmap="YlGnBu", linewidths=1, square=True)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(ticks=range(12), labels=xticks_labels, rotation=90)
plt.xlabel("Month")
plt.show()
```
