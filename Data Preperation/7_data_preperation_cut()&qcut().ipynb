```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/world_population_duplicates.csv", index_col=1)
df
```

```python
df.loc['Uzbekistan']
```

```python
df.loc['India']
```

```python
bins = [0,1000,10000,30000,50000,100000,300000,1500000]
```

```python
population = df.pop2021
population
```

```python
group = pd.cut(population, bins)
group
```

```python
guruh = group.value_counts()
guruh
```

```python
df['guruh'] = group
df
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data.csv", index_col=0)
df2
```

```python
df2.price.describe()
```

```python
bins = [0,10000,20000,50000]
range = pd.cut(df2.price, bins)
range
```

```python
range.value_counts()
```

```python
names = ['arzon', 'o\'rta', 'qimmat']
range = pd.cut(df2.price, bins, labels=names)
range
```

```python
range.value_counts()
```

```python
range = pd.cut(df2.price, 3, labels=names)
range
```

```python
range.value_counts()
```

```python
range = pd.qcut(df2.price, 3, labels=names)
range
```
