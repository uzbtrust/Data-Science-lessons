```python
import pandas as pd
from pandas import DataFrame, Series
import numpy as np
```

```python
df = DataFrame(np.random.randn(4, 3), columns=list('bde'), index=['Utah', 'Ohio', 'Texas', 'Oregon'])
df
```

```python
np.abs(df)
```

```python
f = lambda x: x.max() - x.min()
df.apply(f)
```

```python
def foo(x):
    return Series([x.min(), x.max()], index=['min', 'max'])
df.apply(foo)
```
